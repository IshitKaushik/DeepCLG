```python
import pandas as pd

train = pd.read_csv("dataset/UNSW_NB15_training-set.csv")
test = pd.read_csv("dataset/UNSW_NB15_testing-set.csv")

print("Train shape:", train.shape)
print("Test shape:", test.shape)
print(train.head())
```

```python
print(train.columns)
```

```python
print(train["label"].value_counts())
print(train["attack_cat"].value_counts())
```

```python
# X = features (input)
# y = label (output we want to predict)

X_train = train.drop(["label", "attack_cat"], axis=1)
y_train = train["label"]

X_test = test.drop(["label", "attack_cat"], axis=1)
y_test = test["label"]

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
print(X_train.dtypes)
```

```python
print(X_train.head())
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

X_train_enc = X_train.copy()
X_test_enc = X_test.copy()

cat_cols = ["proto", "service", "state"]

for col in cat_cols:
    le = LabelEncoder()
    
    # combine train + test values
    combined = pd.concat([X_train_enc[col], X_test_enc[col]], axis=0)
    le.fit(combined)

    # transform both
    X_train_enc[col] = le.transform(X_train_enc[col])
    X_test_enc[col] = le.transform(X_test_enc[col])

print("Encoding done successfully âœ…")
print(X_train_enc.dtypes)
print(X_train_enc.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_enc)
X_test_scaled = scaler.transform(X_test_enc)

print("Scaling done âœ…")
print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression(max_iter=1000)

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
